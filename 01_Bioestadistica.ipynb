## Distribución normal  
**Definición**: *la función de densidad* de probabilidad caracteriza el comportamiento probable de un apoblación espiecífica la posabilidad relativa de que una variable continua X tome un valor de x.  
   
La *función de densidad acumulativa* de X es: $$F(x)=\int_{-\infty}^\infty f(x)dx = P (X\leq x)$$

```python
import numpy as np
import math as m
from matplotlib import pyplot as plt
% pylab inline

# Función que con un rango de x, la media y la desviación estandar regresa un vector de probabilidad
def pdf(x,mu,sigma):
    return 1./(np.sqrt(2.*np.pi)*sigma)*np.exp(-np.power((x-mu)/sigma,2.)/2)
def stnormDs(x,mu,sigma):
    pd = np.zeros(len(x))
    for i in range(0,len(x)):
        pd[i] = pdf(x[i],mu,sigma)
    return pd

# Rango de X de -5 a 5
x = np.arange(-10,10,0.1)

plt.plot(x,stnormDs(x,-2,0.5),'g', label=(r'$\mu=-2$ $\sigma=0.5$'))
plt.plot(x,stnormDs(x,0,1),'r', label=(r'$\mu=0$ $\sigma=1$'))
plt.plot(x,stnormDs(x,0,2),'b', label=(r'$\mu=0$ $\sigma=12$'))
plt.plot(x,stnormDs(x,0,4),'y', label=(r'$\mu=0$ $\sigma=4$'))

plt.grid()
plt.ylabel(r'Density $\varphi_{\mu,\sigma^2}(x)$',fontsize=20)
plt.xlabel(r'X',fontsize=14)
plt.title('Probability Density Function',fontsize=20)
plt.legend()
```

La función de densidad de una variable $X$ <font color='red'>con distribución normal </font> y con media $\mu$ y una varianza $\sigma^2$ es:  
\begin{align}
f(x) = \dfrac{1}{\sigma \sqrt{2\pi}} e^{\dfrac{-(x-\mu)^2}{2\sigma^2}}
\end{align}  
Se denota $X\sim \mathcal{N}(\mu,\sigma^2)$. Se dice $X$ se distribuye normalmente.  
Si $\mu=0$ $\sigma^2=1$ se dice que $X\sim \mathcal{N}(0,1)$ tiene distribución normal <font color='red'> estandar </font>.  
Si $X\sim \mathcal{N}(0,1)$ entonces su pdf es:  
\begin{align}
f(x) = \dfrac{1}{\sqrt{2\pi}} e^{\dfrac{-x^2}{2}}
\end{align}  
  
### Propiedades de la PDF  
1. Área total bajo la curva: $$\int_{-\infty}^\infty f(x) dx = 1$$
2. Función negativa: $$f(x)> 0$$  
3. Probabilidad de X en el intervalo ab: $$P(a\leq X\leq b) = \int_a^b f(x)dx$$

### Ejemplo 1
En una ciudad se estima que la temperatura máxima en el mes de junio sigue una distribución normal con $\mu$=23ºC y $\sigma$=5ºC.  Calcular el número de días del mes en que se espera alcanzar máximas entre 21 y 27 grados.  
$\mu=23$, $\sigma=5$  $P(21\leq X\leq 27)$  
$$z_1=\dfrac{x-\mu}{\sigma}=\dfrac{21-23}{5}=-0.4$$  
$$z_2=\dfrac{27-23}{5}=0.8$$

```python
import scipy.stats as st
x = np.arange(3,43,0.1)
plt.plot(x,stnormDs(x,23,5),'b', label=(r'$\mu=0$ $\sigma=2$'))
section = np.arange(21, 27, 0.1)
plt.fill_between(section,pdf(section,23,5))
print  ("P(21< X <27) = %s")%(st.norm.cdf(0.8)-st.norm.cdf(-0.4))
print "Probabilidad de %s días de lluvia"%((st.norm.cdf(0.8)-st.norm.cdf(-0.4))*30)
```

### Ejemplo 2
La media de los pesos de 500 estudiantes es de 70kg y la desviación tipica es de 3 kg. Suponiendo que los pesos se distribuyen normalmente, hay que encontrar cuantos estudiantes pesan:  
a) $P(60\leq X\leq 65)$  
b) $P(X \geq 90)$  
c) $P(X \leq 64)$

```python
#Valores Z Para los estudiantes con P(60>X>65)
z1=(60.-70)/3
z2=(65.-70)/3
P=z2-z1
print "Z de 60 y 65 es entre %s y %s"%(z1, z2)

#Valores Z  Para el estudiante de P(X>90)
z3=(90.-70)/3
print "Z de 90 es %s" %(z3)

#Valores Z Para el estudiante de P(X<64)
z4=(64.-70)/3
print "Z de 64 es %s" %(z4)

# Cumulative Distribution Function CDF(x) = Pr(X<=x) st.norm.cdf
print "P(60>x>65) = %s" %(st.norm.cdf(-1.666666666)-st.norm.cdf(-3.3333333))
print "P(X>90) = %s" %(1-st.norm.cdf(6.66666666667))
print "P(X<64) = %s" %(st.norm.cdf(-2))
```

>Nota, hay que recordar que:  
    $f(x) = \dfrac{1}{\sigma \sqrt{2\pi}} e^{\dfrac{-(x-\mu)^2}{2\sigma^2}}$

```python
rcParams
op = rcParams['figure.figsize']
rcParams['figure.figsize'] = 15,4
def Ycdf(x,mu,sig):
    cdf = np.zeros(len(x))
    for i in range(0,len(x)):
        cdf[i] =  st.norm.cdf((x[i]-mu)/sig)
    return cdf
x = np.arange(60,80,0.1)           #Valores de x, peso en kg
plt.subplot( 1, 2, 1 )
plt.plot(x,stnormDs(x,70,3),'b')   #Gráfica de la funcion de distribucion
plt.fill_between(np.arange(60, 65, 0.1),pdf(np.arange(60, 65, 0.1),70,3)) #Relleno de la integral de interes fill(x,funcion)
plt.xlabel(r'Peso (Kg)',fontsize=14)
plt.text(62, .1, r'$\mu=70,\ \sigma=3$' '\n' r'$P(60\leq X\leq 65)$',fontsize=16)
plt.ylabel(r'$\varphi_{\mu,\sigma^2}(x)$',fontsize=20)
plt.title('Probability Density Function',fontsize=20)

plt.subplot( 1, 2, 2 )
plt.plot(x,Ycdf(x,70,3))
plt.title('Cumulative distribution function',fontsize=20)
plt.xlabel(r'$X$',fontsize=14)
plt.ylabel(r'$\phi_{\mu,\sigma^2}(x)$',fontsize=20)
plt.fill_between(np.arange(60, 65, 0.1),Ycdf(np.arange(60, 65, 0.1),70,3))
```

### dnorm  
The density function of the standard normal distribution is:
\begin{equation}
\phi(x) = (2\pi)^{−1/2} exp(−x^2/2)
\end{equation}
The distribution is called the standard normal because a random variable
X sampled from this distribution will have zero mean and unit variance. This
function is available in `R` as `dnorm()`.  
Regresa el valor de la densidad acumulada de X, dada la media y la sd.
